```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

# Set up Chrome options to run with a visible UI (no headless mode)
chrome_options = Options()
chrome_options.add_argument("--disable-gpu")  # Disable GPU for better performance in some cases

# Set a User-Agent to simulate a real browser request
chrome_options.add_argument("User-agent': 'Mozilla/5.0 (Linux; Android 11; SAMSUNG SM-G973U) AppleWebKit/537.36 (KHTML, like Gecko) SamsungBrowser/14.2 Chrome/87.0.4280.141 Mobile Safari/537.36")

# Initialize WebDriver (update the path to where your chromedriver is located)
driver = webdriver.Chrome(service=Service('/Users/sdarcyrice/Desktop/chromedriver'), options=chrome_options)

# Open the webpage
driver.get("https://www.bershka.com/gb/women/clothes/dresses-n3802.html")

# Let the page load for a few seconds
time.sleep(5)

# Initialize an empty list to store product details
product_list = []

# Set the number of scrolls and the pause time between each scroll
scrolls = 40  # Number of scrolls to simulate (adjust as needed)
scroll_pause_time = 2  # Time to wait after each scroll to let the content load

# Simulate scrolling through the page using JavaScript
for _ in range(scrolls):
    # Execute JavaScript to scroll down by 1000 pixels
    driver.execute_script("window.scrollBy(0, 1000);")
    
    # Wait a little to ensure content is loaded after scroll
    time.sleep(scroll_pause_time)

    # Get the page source after the scroll
    soup = BeautifulSoup(driver.page_source, 'lxml')

    # Find all products in the page source (dresses)
    products = soup.find_all('li', class_='grid-item normal')

    # Extract the details for each product
    for product in products:
        try:
            name = product.find('div', class_='product-text').text.strip()
            price = product.find('span', class_='current-price-elem').text.strip()

            # Append each product's details as a dictionary to the list
            product_details = {
                'name': name,
                'price': price
            }
            product_list.append(product_details)
        except AttributeError:
            continue  # If any data is missing, skip the product

# After scrolling through the page, create a DataFrame from the list of product details
df = pd.DataFrame(product_list)

# Print the DataFrame with all products
print(df)

# Close the browser once you're done
#driver.quit()
```

```python
# drop top two rows that include categories
df['price'] = df['price'].replace('£','',regex=True)
df['price'] = df['price'].replace('From','',regex=True)
df['price'] = df['price'].replace(',','',regex=True)
df['price'] = pd.to_numeric(df['price'])

# Optionally, reset the index after removing duplicates
df = df.reset_index(drop=True)
```
