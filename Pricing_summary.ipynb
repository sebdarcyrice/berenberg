### Histograms

```python
def plot_histogram(bins, df, brand_name, overflow_threshold):
    # Create a new column with prices adjusted for overflow
    df['price_overflow_adjusted'] = np.where(df['price'] > overflow_threshold, overflow_threshold, df['price'])
    
    plt.figure(figsize=(10, 6))
    # Use the adjusted column for plotting
    n, bins, patches = plt.hist(df['price_overflow_adjusted'], bins=bins, edgecolor='black', color='#893B1F')
    plt.title(f'{brand_name}', fontsize=20, fontweight='bold')

    # Remove surrounding box
    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)

    # Format x-axis labels as integers
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}'))
    
    plt.xlabel('Price (£)')
    plt.ylabel('Frequency')
    plt.grid(False)

    # Calculate the maximum price
    max_price = df['price'].max()

    # Add label above the last bar
    last_bar_x = bins[-1]  # Get the x position of the last bar
    last_bar_height = n[-1]  # Get the height (frequency) of the last bar
    label_text = f'Dresses > £{overflow_threshold}'
    max_price_text = f'Max dress price: £{max_price:.2f}'

    # Remove y-axis ticks (but keep the labels)
    ax.yaxis.set_ticks_position('none')
    
    # Place the label above the last bar with two lines
    # Slightly raise the label by adding a small offset to the y position
    vertical_offset = 5
    plt.text(last_bar_x, last_bar_height + vertical_offset, f"{label_text}\n{max_price_text}", 
             ha='center', va='bottom', fontsize=12, color='black', 
             verticalalignment='bottom')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_histogram(ax, bins, df, brand_name, overflow_threshold):
    # Create a new column with prices adjusted for overflow
    df['price_overflow_adjusted'] = np.where(df['price'] > overflow_threshold, overflow_threshold, df['price'])
    
    # Use the adjusted column for plotting
    n, bins, patches = ax.hist(df['price_overflow_adjusted'], bins=bins, edgecolor='black', color='#893B1F')
    ax.set_title(f'{brand_name}', fontsize=10, fontweight='bold')

    # Remove surrounding box
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)

    # Format x-axis labels as integers
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}'))
    ax.grid(False)

    # Calculate the maximum price
    max_price = df['price'].max()

    # Add label above the last bar
    last_bar_x = bins[-1]  # Get the x position of the last bar
    last_bar_height = n[-1]  # Get the height (frequency) of the last bar
    label_text = f'Dresses > £{overflow_threshold}'
    max_price_text = f'Max price: £{max_price:.2f}'

    # Remove y-axis ticks (but keep the labels)
    ax.yaxis.set_ticks_position('none')
    
    # Place the label above the last bar with two lines
    # Slightly raise the label by adding a small offset to the y position
    vertical_offset = 5
    ax.text(last_bar_x, last_bar_height + vertical_offset, f"{label_text}\n{max_price_text}", 
             ha='center', va='bottom', fontsize=8, color='black', 
             verticalalignment='bottom')


# Creating a 3x6 grid of subplots
fig, axs = plt.subplots(6, 3, figsize=(15, 20))

# Plot each histogram on the corresponding axis
plot_histogram(axs[0, 0], 50, df_asos, "ASOS", 160)
plot_histogram(axs[0, 1], 50, df_bershka, "Bershka", 50)
plot_histogram(axs[0, 2], 50, df_boohoo, "Boohoo", 70)

plot_histogram(axs[1, 0], 50, df_cos, "Cos", 200)
plot_histogram(axs[1, 1], 50, df_depop, "Depop", 150)
plot_histogram(axs[1, 2], 50, df_handm, "H&M", 160)

plot_histogram(axs[2, 0], 50, df_massimo, "Massimo Dutti", 190)
plot_histogram(axs[2, 1], 50, df_mands, "M&S", 250)
plot_histogram(axs[2, 2], 50, df_next, "Next", 250)

plot_histogram(axs[3, 0], 50, df_primark, "Primark", 30)
plot_histogram(axs[3, 1], 50, df_pullandbear, "Pull&Bear", 50)
plot_histogram(axs[3, 2], 50, df_shein, "Shein", 25)

plot_histogram(axs[4, 0], 50, df_stradivarius, "Stradivarius", 160)
plot_histogram(axs[4, 1], 50, df_temu, "Temu", 25)
plot_histogram(axs[4, 2], 50, df_vinted, "Vinted", 15)

plot_histogram(axs[5, 0], 50, df_zalando_de, "Zalando DE", 500)
plot_histogram(axs[5, 1], 50, df_zalando_uk, "Zalando UK", 500)
plot_histogram(axs[5, 2], 50, df_zara, "Zara", 100)

# Adjust layout to avoid overlapping labels
plt.tight_layout()

# Show the plot
plt.show()
```
