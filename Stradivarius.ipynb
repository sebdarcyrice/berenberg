```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# Set up Selenium
driver = webdriver.Chrome()  # Make sure you have ChromeDriver installed
url = 'https://www.stradivarius.com/gb/women/clothing/dresses-n1995'
driver.get(url)

# Wait for the products to load (adjust time as needed)
time.sleep(20)

# Retrieve the page source and parse with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
print(soup)
```

```python
# Extract product information from <noscript> or directly from visible HTML
products = soup.find_all('li')
product_list = []

for product in products:
    name_tag = product.find('p')
    price_tag = product.find_all('p')
    
    # Extract name and price
    name = name_tag.get_text(strip=True) if name_tag else None
    price = price_tag[1].get_text(strip=True) if len(price_tag) > 1 else None
    
    if name and price:
        product_list.append({'name': name, 'price': price})

df = pd.DataFrame(product_list)
print(df)

# It works and loads in all of the dresses!!!
```

```python
df.to_excel('stradivarius_dresses.xlsx', index=False) 
```
