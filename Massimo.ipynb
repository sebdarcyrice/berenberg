```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup

# Set up Selenium WebDriver
driver = webdriver.Chrome()  # Ensure ChromeDriver is installed and in PATH
url = 'https://www.massimodutti.com/gb/women/dresses-n1494'
driver.get(url)

# Wait for the page to load initially
time.sleep(5)

# Scroll down the page incrementally
scroll_pause_time = 20  # Adjust if needed
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Wait for the page to load new content
    time.sleep(scroll_pause_time)
    
    # Calculate new scroll height and compare with last height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break  # Exit the loop if the height hasn't changed
    last_height = new_height

# Retrieve the page source and parse it with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Close the browser
driver.quit()

# Continue with your scraping logic here
```

```python
print(soup)
```

```python
# Extract product information from <noscript> or directly from visible HTML
products = soup.find_all('product-view-footer')
product_list = []

for product in products:
    # Check if 'h3' for name exists
    name_tag = product.find('span', class_='cursor-pointer')
    name = name_tag.text.strip() if name_tag else 'No Name'
    
    # Check if 'span' for price exists
    price_tag = product.find('div', class_='text-nowrap title-l')
    price = price_tag.text.strip() if price_tag else 'No Price'
    
    product_details = {
        'name': name,
        'price': price
    }
    
    product_list.append(product_details)

df = pd.DataFrame(product_list)
print(df)
```

```python
# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns

# Now print the DataFrame
print(df)
```

```python
df_dresses = df.head(95)
```

```python
# Remove rows where 'name' is 'No Name' and 'price' is 'No Price'
df_dresses['price'] = df_dresses['price'].replace('£','',regex=True)
df_cleaned = df_dresses[~((df_dresses['name'] == 'No Name') & (df_dresses['price'] == 'No Price'))]

df_cleaned = df_cleaned.reset_index(drop=True)
# Print the cleaned DataFrame
print(df_cleaned)
```
